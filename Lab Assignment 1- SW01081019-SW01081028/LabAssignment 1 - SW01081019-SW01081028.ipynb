```python
import re
import json
import requests
import csv
from datetime import datetime
from bs4 import BeautifulSoup

url ='https://shopee.com.my/Logitech-G-PRO-X-Superlight-Logitech-G-PRO-X-Superlight-2-Wireless-Gaming-Mouse-i.58359596.3876090618?is_from_login=true'

r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

offset = 0
reviews = []  # Initialize an empty list to store reviews

while True:
    ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
    data = requests.get(ratings_url).json()

    for rating in data['data']['ratings']:
        author_username = rating['author_username']
        comment = rating['comment']
        timestamp = rating['ctime']
        readable_time = datetime.fromtimestamp(timestamp).strftime('%y-%m-%d %H:%M:%S')

        reviews.append({
            'author_username': author_username,
            'date': readable_time,  # Changed 'Date' to 'date' for consistency
            'comment': comment,
        })

    if len(data['data']['ratings']) < 20:
        break

    offset += 20

# Save contents to CSV file
with open('content3.csv', 'w', encoding='utf-8', newline='') as csvfile:
    fieldnames = ['author_username', 'comment', 'date']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()  # Write Headers

    for review in reviews:
        writer.writerow(review)

print("Reviews have been saved to content3.csv")
```
